```python
from backtesting import *
import numpy as np
```

```python
# Instantiate backtesting
model = Backtest()
```

```python
# Test scores for sipple multivariable regression (with some rough variable selection)
features = [
#     'Cumulative cases',
#     'Cumulative Vancouver Coastal',
    'Hospitalization (Currently in hospital)',
    'ICU (Currently in ICU)',
    'Recovered',
    'Deaths',
#     'Active cases',
#     'Annacis Island'
]

for day in range(7,35):
  RMSPE = model(output_predication_csv=False, model='regression', days=day, features=features)
  print(f'day: {day} RMSPE: {RMSPE}')
```

```python
for day in range(15,35):
    for alpha in np.linspace(0,1,15):
        RMSPE = model(output_predication_csv=False, alpha=alpha, model='ridge_regression', days=day, features=features)
        print(f'day: {day} alpha: {alpha} RMSPE: {RMSPE}')
```
